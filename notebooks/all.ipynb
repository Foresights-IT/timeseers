```python
%load_ext autoreload
%autoreload 2

from seers import utils, LinearTrend, TimeSeriesModel
import pandas as pd
import numpy as np
```

```python
data = utils.generate_data()

X = pd.DataFrame({'t': np.arange(len(data))})
y = pd.Series(data)
```

```python
model = TimeSeriesModel() + LinearTrend(n_changepoints=24)
model.fit(X, y)
```
